```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

# Data gathering by BeatifulSoup

Almaty district(Astana)

```python
# page which we want to parse
url = 'https://www.kn.kz/astana/prodazha-kvartir/?region%5Bvalues%5D=31274108'
headers = {'user-agent': 'Mozilla/5.0'}
response = requests.get(url, headers=headers)
html = response.text
soup = BeautifulSoup(html, 'html.parser')
```

```python
# finding number of pages
pagination = soup.find('div',attrs={'class':'paginator'}).find_all('span')
pages = pagination[-2].text
print('Всего страниц: ' + pages)
```

https://www.kn.kz/astana/prodazha-kvartir/page/2/?region%5Bvalues%5D=31274108

```python
# getting data from every page
houses = []
for page in range(1, int(pages)+1):
    response = requests.get("https://www.kn.kz/astana/prodazha-kvartir/"+'page/'+str(page)+"?region%5Bvalues%5D=31274108")
    html = response.text
    soup = BeautifulSoup(html, 'html.parser')
    table = soup.find('ul', attrs = {'class':'results-list'})
    print(f'Parsing of page {page} from {pages}...')
    for row in table.findAll('div',attrs = {'class':'object-row-data'}):
            house = {}
            house['rooms_and_street'] = row.find('a', attrs = {'class':'results-item-street'}).text
            house['num_of_floors'] = row.find('li', attrs = {'class':'results-item-floor item field field-floor'}).text
            house['sqrft'] = row.find('li', attrs = {'class':'results-item-square item field field-area-value'}).text
            house['year_of_build'] = row.find('li', attrs = {'class':'results-item-build item field field-build'}).text
            house['description'] = row.find('p', attrs = {'class':'results-item-descr'}).text
            house['date'] = row.find('span', attrs = {'class':'results-item-date'}).text
            house['rieltor'] = row.find('div', attrs = {'class':'object-author'}).text
            house['price'] = row.find('p', attrs = {'class':'results-item-price'}).text
            houses.append(house)
```

```python
import pandas as pd
df_almaty = pd.DataFrame(houses)
```

```python
df_almaty.head()
```
